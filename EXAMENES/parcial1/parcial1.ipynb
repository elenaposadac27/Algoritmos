#PUNTO 1

```
import requests
url = "https://listado.mercadolibre.com.co/gatos-siames#D[A:gatos%20siames]"

response = requests.get(url)
print(response)
```

```
import bs4
completo = bs4.BeautifulSoup(response.text, "html.parser")
dataEspecificacion =completo.select(".ui-search-item__title")
dataPrecio = completo.select(".price-tag-fraction")
print(dataEspecificacion)
print(dataPrecio)

especificacionGatos = []
preciosGatos = []

funcionExtraerTexto = lambda elemento : elemento.text

especificacionGatos = list(map(funcionExtraerTexto, dataEspecificacion))
preciosGatos = list(map(funcionExtraerTexto, dataPrecio))
print(especificacionGatos)
print(preciosGatos)
```

#PUNTO 2 Y 3

```
import pandas as pd

print(len(especificacionGatos))
print(len(preciosGatos))

dictGatos = {}
dictGatos["Siames"] = especificacionGatos
dictGatos["Precios"] = preciosGatos [:46]

siamesDataFrame = pd.DataFrame(dictGatos)
siamesDataFrame.to_csv("Gatos Siames.csv", index=False)
```

#PUNTO 4

```
funcionAgregarGato = lambda elemento : "Gato:" + elemento
titulosGatos = list(map(funcionAgregarGato,especificacionGatos))

dictGatos = {}
dictGatos["Siames"] = titulosGatos
dictGatos["Precios"] = preciosGatos [:46]

gatosDataFrame = pd.DataFrame(dictGatos)
gatosDataFrame.to_csv("NEW Gatos Siames.csv", index=False)
```

#PUNTO 5

```
snowshoe = lambda elemento: "Snow" in elemento
listaSnowshoe = list (filter(snowshoe, especificacionGatos))
print(listaSnowshoe)

snowshoeSerie = pd.Series(listaSnowshoe)
snowshoeSerie.to_csv("Gatos Siames SNOWSHOE  .csv", index=False)

```

```
menores80 = lambda elemento : len(elemento) < 80
listaTitulosMenores80 = list(filter(menores80,listaSnowshoe))
print(listaTitulosMenores80)

snowshoeSerie80 = pd.Series(listaTitulosMenores80)
snowshoeSerie80.to_csv("Gatos Siames SNOWSHOE (titulos menores 80).csv", index=False)
```

#PUNTO 6

```
from functools import reduce

largosEspecificacionGatos = []
for elemento in especificacionGatos:
  largosEspecificacionGatos.append(len(elemento))
print(largosEspecificacionGatos)

acumulador= 0
largo = lambda acumulador =0 , elemento =0 : acumulador + elemento
largoPromedio = reduce (largo, largosEspecificacionGatos)/len(largosEspecificacionGatos)
print(largoPromedio)
```
