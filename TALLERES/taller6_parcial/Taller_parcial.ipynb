#PUNTO 1 Y 2


```
import requests
url = "https://listado.mercadolibre.com.co/cachorros-bulldog-frances#D[A:cachorros%20bulldog%20frances]"

response = requests.get(url)
print(response)
```

```
import bs4
completo = bs4.BeautifulSoup(response.text, "html.parser")
dataEspecificacion =completo.select(".ui-search-item__title")
dataPrecio = completo.select(".price-tag-fraction")
print(dataEspecificacion)
print(dataPrecio)

especificacionBulldog = []
preciosBulldog = []

funcionExtraerTexto = lambda elemento : elemento.text

especificacionBulldog = list(map(funcionExtraerTexto, dataEspecificacion))
preciosBulldog = list(map(funcionExtraerTexto, dataPrecio))
print(especificacionBulldog)
print(preciosBulldog)
```

#PUNTO 3

```
import pandas as pd
dictBulldog = {}
dictBulldog ["Cachorros"] = especificacionBulldog
dictBulldog ["Precios"] = preciosBulldog[:50]

bulldogDataFrame =pd.DataFrame(dictBulldog)
bulldogDataFrame.to_csv("Cachorros bulldog.csv", index=False)
```

#PUNTO 4

```
funcionAgregarCachorro = lambda elemento : "Cachorro:" + elemento
titulosCachorros = list(map(funcionAgregarCachorro,especificacionBulldog))

dictCachorros = {}
dictCachorros["Cachorros"] = titulosCachorros
dictCachorros["Precios"] = preciosBulldog[:50]

cachorrosDataFrame = pd.DataFrame(dictCachorros)
cachorrosDataFrame.to_csv("NEW Cachorros bulldog.csv", index= False)
```

#PUNTO 5

```
merle = lambda elemento: "Merle" in elemento
listaMerle = list (filter(merle, especificacionBulldog))
print(listaMerle)

merleSerie = pd.Series(listaMerle)
merleSerie.to_csv("Cachorros Bulldog MERLE .csv", index=False)
```

#PUNTO 6

```
from functools import reduce
acumulador = 0

sumatoria = lambda acumulador =0 , elemento =0 : acumulador + elemento
resultado = reduce (sumatoria, especificacionBulldog)
print (resultado)
```

#PUNTO 7

```
menores25 = lambda elemento : len(elemento) < 25
listaTitulosMenores25 = list(filter(menores25,especificacionBulldog))
print(listaTitulosMenores25)
```

#PUNTO 8

```
largosEspecificacionBulldog = []
for elemento in especificacionBulldog:
  largosEspecificacionBulldog.append(len(elemento))
print(largosEspecificacionBulldog)

acumulador= 0
largo = lambda acumulador =0 , elemento =0 : acumulador + elemento
largoPromedio = reduce (largo, largosEspecificacionBulldog)/len(largosEspecificacionBulldog)
print(largoPromedio)
```
