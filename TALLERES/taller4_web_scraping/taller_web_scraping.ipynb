```
import requests
import bs4

url = "https://www.semana.com/"

response = requests.get(url)
print(response)
```

```
completo = bs4.BeautifulSoup(response.text,'html.parser')
noticias = completo.select(".article a")
print (noticias)
```

```
titulos =[]
links = []
for noticia in noticias:
  try:
    links.append('https://www.semana.com/'+ noticia['href'])
    titulos.append(noticia.text)
  except:
    print("articulo_feo")

print (titulos)
print (links)
dictNoticias = {}
dictNoticias['Noticias']= titulos
dictNoticias['links']= links
```

```
import pandas as pd 
noticiasSemana = pd.DataFrame(dictNoticias)
noticiasSemana

noticiasSemana.to_csv('noticiasSemana.csv', index=False)
```
